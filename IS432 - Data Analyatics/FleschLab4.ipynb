```python
import matplotlib.pyplot as plt
df = pd.read_csv("bankfull.csv")
histo_d = df.hist(['pdays'], bins = 10)
plt.ylabel('Frequencies')
plt.xlabel('Last time a client called')
```

<h3> We can observe from this histogram that there is a significant difference between the number of days customers were last contacted, which is from day 1 to day 999, and that the days clients were last phoned was after 999 days which has a higher frequency than those called lately.<h3/>

```python
import numpy as np
df['pdays']=df['pdays'].replace({999: np.NaN})
histo_d = df.hist(['pdays'], bins = 10)
plt.ylabel('Frequencies')
plt.xlabel('Last time a client called')
```

<h3> After creating another histogram, we can notice that the data is much closer compared to the old historagam. We can also see that there is a bigger number of customers who have not recently returned calls and can see the data that is relevant to the campaign on who called back, this data shows an cleaner histogram.<h3/>

```python
from scipy import stats
df['age_z']=stats.zscore(df['age'])
```

```python
df['age_z'].hist()
plt.ylabel('Client')
plt.xlabel('Zscore') 
plt.title('Age')
```

<h3><h3/>

```python
df_outliers = df.query('age_z > 3 | age_z < -3')
```

```python
s_out = np.array([df_outliers])
print (np.shape(s_out))
```
