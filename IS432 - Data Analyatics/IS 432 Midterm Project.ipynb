Because the state indicated by NAN was only loosely described, I decided to replace it with Not Available in the missing values. I believe this was the most accurate approach to express the missing value in my dataset.Â 

```python
import pandas as pd
df = pd.read_csv("sales_data_sample.csv")
STATUS = {'Shipped': 0,'Disputed': 1,'In Process': 2, 'Cancelled': 3, 'On Hold': 4, 'Resolved': 5}
df.STATUS = [STATUS[item] for item in df.STATUS]
print(df)
```

Here i've turned categorical variable into a quantitative one, for attribue order status, categorical variable like Shipped, Disputed, In Process, Cancelled, On Hold, Resolved were used initailly which are replace by numerical variables 0, 1, 2, 3, 4, 5. By replacing categorical with quantitative variable it thas given the order status attribute  greater statistical power. 

# Modeling :

```python
import pandas as pd
from scipy import stats
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import pylab
from sklearn.preprocessing import StandardScaler, normalize
from sklearn.decomposition import PCA
import scipy.cluster.hierarchy as hc
from sklearn.cluster import AgglomerativeClustering
df
```

```python
df[['ORDERNUMBER','QUANTITYORDERED','PRICEEACH','ORDERLINENUMBER','SALES','QTR_ID','MONTH_ID','YEAR_ID','MSRP']] = df[[ 'ORDERNUMBER','QUANTITYORDERED','PRICEEACH','ORDERLINENUMBER','SALES','QTR_ID','MONTH_ID','YEAR_ID','MSRP']].apply(pd.to_numeric, errors='coerce')
df = df.dropna()
df = df.reset_index(drop=True)
df = df.dropna()
df = df.reset_index(drop=True)
df
```

```python
Y = df[['QUANTITYORDERED', 'PRICEEACH']]
Yz = pd.DataFrame(stats.zscore(Y), columns = ['QUANTITYORDERED', 'PRICEEACH'])
kmeans01 = KMeans(n_clusters = 2).fit(Yz)
labels = kmeans01.labels_
df["labels"] = labels
df.head()
filtered_label0 = df[labels == 0]
filtered_label1 = df[labels == 1]
plt.scatter(filtered_label0['QUANTITYORDERED'],filtered_label0['PRICEEACH'], color = 'red')
plt.scatter(filtered_label1['QUANTITYORDERED'],filtered_label1['PRICEEACH'], color = 'black')
plt.title('k-means Clustering by Quantity Ordered and Each Price') #title
plt.xlabel('Quantity Ordered')
plt.ylabel('Each Price') 
plt.show()
```

Choose to use K-means clustering to model my dataset. I'v used Price of each unit and Quantity order to represent in my model. I found that K-means model used for my sales dataset to be practical working well with my dataset, it was simple, easy to implement and easy to interpret the clustering results. We can interpret that max number of order price was within the range of 30 - 80 with most order quantity being in the range of 20-50. 

```python
epsilon = 0.5
minimumSamples = 5
Y = df[['QUANTITYORDERED',  'PRICEEACH']]
Yz = pd.DataFrame(stats.zscore(Y), columns = ['QUANTITYORDERED',  'PRICEEACH'])
db = DBSCAN(eps=epsilon, min_samples=minimumSamples).fit(Yz)
labels_db = db.labels_
labels_db
df["labels_db"] = labels_db
filtered_label0 = df[labels_db == 0]
filtered_label1 = df[labels_db == 1]
filtered_label2 = df[labels_db == 2]
filtered_label3 = df[labels_db == 3]
filtered_label4 = df[labels_db == 4]
plt.scatter(filtered_label0['QUANTITYORDERED'],filtered_label0['PRICEEACH'], color = 'red')
plt.scatter(filtered_label1['QUANTITYORDERED'],filtered_label1['PRICEEACH'], color = 'black')
plt.scatter(filtered_label2['QUANTITYORDERED'],filtered_label2['PRICEEACH'], color = 'blue')
plt.scatter(filtered_label3['QUANTITYORDERED'],filtered_label3['PRICEEACH'], color = 'green')
plt.scatter(filtered_label4['QUANTITYORDERED'],filtered_label4['PRICEEACH'], color = 'pink')
plt.title('DBSCAN Clustering by Quantity Ordered and Price Each')
plt.xlabel('Quantity Ordered')
plt.ylabel('Price Each')
plt.show()
```

Choose to use DBSCAN clustering to model my dataset. I've used Price of each unit and Quantity order to represent in my model. DBSCAN was utilized in this and is clear sales data is seen as more simple and it has produced a varying number of clusters, grouping  based on the input data. We can also see that DBScan model method has located regions of high density around 100 Price for each unit and has separated other regions of low density.
